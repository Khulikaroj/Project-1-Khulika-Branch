# 2. Data Access and Cleaning

## Download index for SPY and Factors ETF (Value, Momentum, Quality, Growth/High Dividend, Low Vol)

```python
#download index price
stocks = ["SPY","IVE","SCHV","MTUM","QUAL","IVW","SDIV","HYG","JNK","FDL"]
end_date = datetime.now()
start_date = end_date + timedelta(-365 * 20)
data = yf.download(stocks, start=start_date, end=end_date)
data.tail()
```

```python
#funds dataframe
funds_df = data['Close'].dropna()
funds_df.head()
```

```python
#return dataframe
funds_return = funds_df.pct_change().dropna()
funds_return.columns = ["HD-FDL","HD-HYG","VL-IVE","GR-IVW","HD-JNK","MT-MTUM","QL-QUAL","VL-SCHV","HD-SDIV","SPY"]
funds_return.tail(10)
```

# 3. Predetermine Portfolios and Backtesting

## Analyse historical risk/return of passive vs active funds

### We are looking at return and risk characteristic of S&P index, a passive fund, and active funds like Value funds (VL), Momentum funds (MT), Qyality fund (QL), Growth (GR), and High Dividend (HD). Price and return data are daily frequency covering 07/18/2013 to the most recent date. 

```python
#reorder the columns of price dataframe into fund groups
funds_df.columns = ["HD-FDL","HD-HYG","VL-IVE","GR-IVW","HD-JNK","MT-MTUM","QL-QUAL","VL-SCHV","HD-SDIV","SPY"]
funds_df = funds_df[["VL-IVE","VL-SCHV","MT-MTUM","QL-QUAL","GR-IVW","HD-FDL","HD-HYG","HD-JNK","HD-SDIV","SPY"]]
funds_df.head()
```

```python
#reorder the columns of return dataframe into fund groups
funds_return = funds_return[["VL-IVE","VL-SCHV","MT-MTUM","QL-QUAL","GR-IVW","HD-FDL","HD-HYG","HD-JNK","HD-SDIV","SPY"]]
funds_return.head()
```

### Looking at return distribution in the past 7 years, daily returns of HD-HYG and HD-JNK are less volatile than SPY and other active funds. Volatility of these 2 funds are almost half of SPY. Return volatility of the rest of active funds are similar to S&P index, sightly higher in Momentum fund (MT-MTUM) and Gorwth fund (GE-IVW).

```python
#descriptive statistics
funds_return.describe()
```

```python
#distribution of returns
boxplot_return_dist = funds_return.boxplot(figsize = (15,7), grid = True)
```

```python
#long-term volatilityof return: Why vol in HD-HYG and HD-JNK are lower? 
funds_return_std = funds_return.std()
funds_return_std.plot(kind="bar")
```

### Return of every active funds has positive correlation to S&P. However,HD-HYG and HD-JNK are less correlated to S&P index. Meanwhile, return of Quality fund (QL-QUAL) and Growth fund (GR-IVW) are most correlated to S&P.

```python
# return correlation, beta: positive correlation. High Dividend EFTs seems to correlated more 
correlation = funds_return.corr()
sns.heatmap(correlation, vmin =-1, vmax=1)
```

### In performance space, not all active funds that outperform S&P in the past 7 years. Only Momentum fund (MT-MTUM), Growth fund (GR_IVW) and Quality fund (QL-QUAL) that outperform S&P through out the past 7 years. Value funds (VL-IVE, VL-SCHV), High Dividend funds (HD) underperform S&P. HD-SDIV is the worst performer and started to lag behind other funds in 2016.

```python
# Cummulative return: Growth, Momentum and Qualirty ETFs outperform SPY during 2013-current.
funds_return_cum = (1+funds_return).cumprod()
funds_return_cum.plot(figsize = (15,7), title = "ETFs Cummulative Returns During 2013-current")
```

### Annaulize return from investing in S&P in the past 7 years is 10.29% (annualized return since inception). The best performer among active funds is Momentum (MT-MTUM) which generate annualized return at 14.60%. Follow by Growth (GR-IVW, 14.37%), Quality (QL-QUAL, 10.91%). Even Value funds underperform S&P but annualized return are still positive. Majority of High Dividend funds generate loss.

### However, rank of performance change in recent shorter horizon. In the past 3 months, only Growth fund (GR-IVW, 20.91%) and a High Dividend funds (HD-SDIV, 18.07%) that outperform S&P (17.71%). Outperformance of Growth fund (GR-IVW) over market index is pretty robust when consider different investment horizon (6m, YTD, 1Y-5Y).

### Growth fund is also perform slightly better than market index during tough time. Maximum Drawdown of Momentum fund is -34.08%, slightly than -34.10% of S&P index. However, HDD is more shallow for HD-HYG (-28.05%). In contrast, MDD are deeper in HD-FDL (-41.40%) and HD-SDIV (-69.08%).

### Momentum, Quality and Growth funds also have more upside return, less downside return compared to S&P. (Avg. Up Month, Avg. Down Month). Less upside, more downside are seen in Value fund and some High Dividend funds like HD-FDL and HD-SDIV. This return characteristic makes these funds underperform market index. In HD-HYG and HD-JNK, underperformance came from less upside return in good months even these fund perform better tha S&P in bad months. Such characteristics reflect in monthly volatility which are lower than S&P. Lower vol seems perfect but cause these two portfolios fail to beat merkat index.

```python
#backtesting
backtest = funds_df.calc_stats()
backtest.display()
```

```python
#short-term volatiliry of return: peaked during COVID-19 pandemic 
funds_return_std_roll = funds_return.rolling(window=20).std()
funds_return_std_roll.plot(figsize = (15,7))
```

### In term of tradeoff between risk and return, Momentum, Quality and Growth funds offer better risk-return tradeoff compared to S&P index.

```python
# Sharpe ratio: Risk/return tradeoff are less attractive in High Dividend funds. Most attractive in Momentum, Quality and Growth funds.
funds_return_sharpe = round((funds_return.mean()*252)/(funds_return.std() * np.sqrt(252)),2)
funds_return_sharpe.plot(kind="bar")
```

## Examine risk and return characteristics of passiv and active funds in different economic/market scenarios in the past to get better idea how Momentum, Quality and Growth funds outperform market index.

```python
#economic cycle:cutting cycle(none),Taper Tantrum(2013), goldilock/hike cycle(2016-18), pandemic(Mar-May2020)
#slice return datafram into scenarios
funds_return_taper_tantrum = funds_return.loc['20130718':'20131231']
funds_return_hike_cycle = funds_return.loc['20160102':'20181231']
funds_return_pandemic = funds_return.loc['20200302':'20200330']
#slice price dataframe into scenarios
funds_df_taper_tantrum = funds_df.loc['20130718':'20131231']
funds_df_hike_cycle = funds_df.loc['20160102':'20181231']
funds_df_pandemic = funds_df.loc['20200302':'20200330']
```

### Interestingly, every funds correalted more with S&P during crisis time when volatiliry in the market spike. 

```python
#correlation in different periods/cycle: pandemic: In crisis, correlation of all funds increased to almost 1!
correlation_pandemic = funds_return_pandemic.corr()
sns.heatmap(correlation_pandemic, vmin =-1, vmax=1)
```

```python
#backtest funds during Taper Tantrum
backtest_taper_tantrum = funds_df_taper_tantrum.calc_stats()
backtest_taper_tantrum.display()
```

```python
#backtest funds during hiking cycle
backtest_hike_cycle = funds_df_hike_cycle.calc_stats()
backtest_hike_cycle.display()
```

```python
#backtest funds during pandemic
backtest_pandemic = funds_df_pandemic.calc_stats()
backtest_pandemic.display()
```
